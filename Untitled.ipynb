```python
from pulp import *

# Initialize the problem
model = LpProblem("Product Mix Optimization", LpMaximize)

# Decision variables
A = LpVariable("Product A", lowBound=0, cat='Continuous')
B = LpVariable("Product B", lowBound=0, cat='Continuous')

# Objective Function (Profit per unit: A = $20, B = $30)
model += 20 * A + 30 * B, "Total Profit"

# Constraints
model += 2 * A +  3* B <= 100, "Machine M1 hours"
model += 1 * A + 1 * B <= 80, "Machine M2 hours"
# Solve the model
model.solve()
print(f"Status: {LpStatus[model.status]}")
print(f"Optimal number of Product A to produce: {A.varValue}")
print(f"Optimal number of Product B to produce: {B.varValue}")
print(f"Maximum Profit: ${value(model.objective)}")
import matplotlib.pyplot as plt

# Visualize decision
labels = ['Product A', 'Product B']
values = [A.varValue, B.varValue]
plt.bar(labels, values, color=['skyblue', 'orange'])
plt.title('Optimal Production Mix')
plt.ylabel('Units')
plt.show()


```
